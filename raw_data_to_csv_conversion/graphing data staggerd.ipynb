```python
#combine county data by date, but this time staggered for one week
from datetime import datetime, timedelta
county_date_staggered={}

for i in counties:
    county_date_staggered[i]=[]
    #case and weather iterates through each respictive row
    for case in county_cases[i]:
        virus_date=datetime(int(case[0].split("-")[0]),int(case[0].split("-")[1]),int(case[0].split("-")[2]))
        weather_date=virus_date - timedelta(days=7)
        for weather in county_weather[i]:
            if weather[1] == "Date":
                continue
            weather_date_temp=datetime(int(weather[1].split("-")[0]),int(weather[1].split("-")[1]),int(weather[1].split("-")[2]))
            #looking for dates one week apart from each other, with the weather being the bottom bitch
#             print("two apposing forces")
#             print(virus_date)
#             print(weather_date)
            
            
            if weather_date == weather_date_temp:
                #add an array that connects weather and case data
                county_date_staggered[i].append([case,weather])
```

# the below cell makes csv's for the staggered data. only run it once.

```python
staggered_header = ['Virus Date','County','State','Total Cases','Deaths','Station ID','Weather Date','Latitude','Longitude','Elevation','(file)name',
                      'Temperature','Temp. Attributes','Windspeed','Rain','Rain Attributes',]
for filename in county_date_staggered:
    for row in county_date_staggered[filename]:
        print(row[0],row[1])

#this makes the csv files with only the info we will use for testing.
#only run this once and make sure it spits out a correct csv file.
directory=r"cs175-final-project-master\staggered_virus_and_weather_data"

for filename in county_date_staggered:
    with open(directory+"\\"+filename,'w') as csvfile:
        filewriter = csv.writer(csvfile, delimiter=',',lineterminator = '\n')
        #csvfile=county_weather[filename]
        filewriter.writerow(staggered_header)
        for row in county_date_staggered[filename]:
            filewriter.writerow([row[0][0],row[0][1],row[0][2],row[0][3],row[0][4],
                                row[1][0],row[1][1],row[1][2],row[1][3],
                                row[1][4],row[1][5],row[1][6],row[1][7],
                                row[1][8],row[1][9],row[1][10]])
```

```python
#finally making the fucking plots
from sklearn.linear_model import LinearRegression
linear_regressor = LinearRegression()

for filename in county_date_staggered:
    fig, plotting = plt.subplots(1, 1, figsize=(7, 7))
    plotting.set_title("Relationship between weather and cases, with weather staggered by one week")
    plt_data=[]
    for i in county_date_staggered[filename]:
        #print(i)
        plt_data.append([i[0][3],i[1][6]])



    plt_data=np.array(plt_data)
    #print(plt_data[:,0],plt_data[:,1])

    X = plt_data[:,0].reshape(-1, 1)  # values converts it into a numpy array
    Y = plt_data[:,1].reshape(-1, 1)  # -1 means that calculate the dimension of rows, but have 1 column
    linear_regressor = LinearRegression()  # create object for the class
    linear_regressor.fit(X, Y)  # perform linear regression
    prediction = linear_regressor.predict(X)  # make predictions


    # linear_regressor.fit(plt_data[:,0],plt_data[:,1])  # perform linear regression
    # prediction = linear_regressor.predict(plt_data[:,0])  # make predictions
    plt.plot(plt_data[:,0],prediction)

    plotting.scatter( plt_data[:,0],plt_data[:,1],color="blue", label=filename)
    plt.xlabel("Total Number cases")
    plt.ylabel("Temperature (F)")
    plotting.legend()
    county_name=filename.split('_')[0]
    plt.savefig("cs175-final-project-master\graphs\Total Number of Cases_week_stagger\\"+county_name+".jpg")
    plt.show()
```
